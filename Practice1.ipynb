## Условие 2:

На складе лежат разные фрукты в разном количестве.\
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе

Пример работы функции:\
`total_fruits(banana=5, mango=7, apple=8)`\
`20`

```python
def total_fruits(**kwargs):
    # принимаем неограниченное количество аргументов в формате фрукт=кол
    total = sum([i for i in kwargs.values()])
    return total

total_fruits(banana=5, mango=7, apple=8)
```

Используйте list comprehensions

```python
advertising_costs = [100, 125, -90, 345, 655, -1, 0, 200]
total_advertising_costs = sum([i for i in advertising_costs if i > 0])
total_advertising_costs
```

### 4.2 Найдите выручку компании в зависимости от месяца

Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.

Используйте аннотирование типов.

```python
date_of_buy = ['2021-09-14', '2021-12-15', '2021-09-08', '2021-12-05', '2021-10-09', '2021-09-30', '2021-12-22', '2021-11-29', '2021-12-24', '2021-11-26', '2021-10-27', '2021-12-18', '2021-11-09', '2021-11-23', '2021-09-27', '2021-10-02', '2021-12-27', '2021-09-20', '2021-12-13', '2021-11-01', '2021-11-09', '2021-12-06', '2021-12-08', '2021-10-09', '2021-10-31', '2021-09-30', '2021-11-09', '2021-12-13', '2021-10-26', '2021-12-09']
amount_of_buy = [1270, 8413, 9028, 3703, 5739, 4095, 295, 4944, 5723, 3701, 4471, 651, 7037, 4274, 6275, 4988, 6930, 2971, 6592, 2004, 2822, 519, 3406, 2732, 5015, 2008, 316, 6333, 5700, 2887]

def total_of_buy_month(date_of: list, amount_of: list) -> dict:
    result = {}
    for i in range(len(date_of)):
        if date_of[i][5:7] in result.keys():
            result[date_of[i][5:7]] = result[date_of[i][5:7]] + amount_of[i]
        else:
            result[date_of[i][5:7]] = amount_of[i]
    return result
total_of_buy_month(date_of_buy, amount_of_buy)
```
